#### 2 - Line Plots

```python
import plotly.graph_objects as go
df_stocks = px.data.stocks()   # you can use many other plotly example datasets
df_stocks.head()
df_stocks.set_index('date', inplace=True)   # set date as index

"""
px.line(
    df_stocks, x=df_stocks.index, y=['GOOG', 'AAPL'],
    labels={'x': 'Date', 'y': 'Price'},
    title='Stock Prices - Apple vs Google'
)
"""

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.index, y=df_stocks.AAPL, 
                         mode='lines', name='Apple'))
fig.add_trace(go.Scatter(x=df_stocks.index, y=df_stocks.AMZN, 
                         mode='lines+markers', name='Amazon'))
fig.add_trace(go.Scatter(x=df_stocks.index, y=df_stocks.GOOG, 
                         mode='lines+markers', name='Google',
                         line=dict(color='firebrick', width=2, dash='dashdot')))
fig.update_layout(title='Stock Prices 2018-2020',
                  xaxis_title='Date', yaxis_title='Price')
```
