### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(x):
  if x > 17:
    return "yes"
  return "no"

df['legal_drinker'] = df['age'].apply(majority)
```

```python
# or I could use a map function
df['legal_drinker2'] = map(majority, df['age'])
```

```python
# Check if it works:
print("{0}, {1}".format(df.loc[df['age'] > 17 ,'legal_drinker'].unique(), df.loc[df['age'] <= 17, 'legal_drinker'].unique()))
```

```python
print("{0}, {1}".format(df.loc[df['age'] > 17 ,'legal_drinker2'].unique(), df.loc[df['age'] <= 17, 'legal_drinker2'].unique()))
```

```python
# check if legal_drinker columns are equal:
df['legal_drinker'].equals(df['legal_drinker2'])
```
