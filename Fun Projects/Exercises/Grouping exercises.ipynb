### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(url, sep="|")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.agg('mean')
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['binary_gender'] = np.where(users['gender'] == 'M', 1, 0)
res1 = ((users.groupby('occupation').binary_gender.agg('sum') / users['occupation'].value_counts())*100).sort_values(ascending=False)
res1
```

or

```python
male_count = users[users['gender'] == 'M'].groupby(['occupation']).agg({'gender': 'count'})
res2 = ((male_count / users.groupby('occupation').agg('count'))['gender'] * 100).sort_values(ascending=False)
res1.equals(res2)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).agg({'age': 'mean'})
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_count = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occupation_count = users.groupby('occupation').count()
((gender_count.div(occupation_count, level='occupation'))['gender']) * 100
```
